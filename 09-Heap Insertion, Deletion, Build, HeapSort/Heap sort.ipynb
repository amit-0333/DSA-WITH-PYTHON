**Heapify**

```python
# left = 2*i + 1
# right = 2*i + 2
# parent = (i-1)//2
#smallest = i

def heapify(arr, n, i):
    smallest = i
    left = 2*i + 1
    right = 2*i + 2

    if left < n and arr[left] < arr[smallest]: #smallest = i
        smallest = left

    if right < n and arr[right] < arr[smallest]:
        smallest = right
        
    if smallest != i:                                 #If parent is not smallest:
        arr[i], arr[smallest] = arr[smallest], arr[i] #Swap parent with smallest child
        heapify(arr, n, smallest)                     #Fix heap again from new position
```

**Build Minheap**

```python
def build_min_heap(arr):
    n = len(arr)
    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i)
```

**Insertion in Minheap**

```python
def insert(heap, value):
    heap.append(value)
    i = len(heap) - 1 #Start From Inserted Index

    while i > 0:
        parent = (i - 1) // 2 
        if heap[parent] > heap[i]:
            heap[parent], heap[i] = heap[i], heap[parent]
            i = parent
        else:
            break
# Insert at end
# While parent is bigger:
#     Swap upward
# Stop when correct position found
```

**Deletion**

```python
def delete_min(heap):
    if len(heap) == 0:
        return None

    root = heap[0]
    heap[0] = heap[-1]
    heap.pop()

    heapify(heap, len(heap), 0)
    return root

# Save root
# Move last element to root
# Remove last
# Fix heap downward
# Return saved root
```

**Heap Sort**

```python
def heap_sort(arr):
    build_min_heap(arr)
    result = []

    while len(arr) > 0:
        result.append(delete_min(arr))

    return result
```

```python
arr = [10, 4, 7, 1, 3]

build_min_heap(arr)
print("Min Heap:", arr)

insert(arr, 2)
print("After Insert:", arr)

print("Deleted Min:", delete_min(arr))
print("After Deletion:", arr)

print("Heap Sort:", heap_sort(arr.copy()))
```

**Using heapq**

```python
import heapq

arr = [10, 4, 7, 1, 3]

heapq.heapify(arr)   # Converts list into Min Heap

print("Min Heap:", arr)
```

```python
heapq.heappush(arr, 2)
print("After Insert:", arr)
```

```python
smallest = heapq.heappop(arr)
print("Deleted Min:", smallest)
print("After Deletion:", arr)
```

```python
def heap_sort(nums):
    heapq.heapify(nums)
    sorted_list = []

    while nums:
        sorted_list.append(heapq.heappop(nums))

    return sorted_list


arr = [10, 4, 7, 1, 3]
print("Heap Sort:", heap_sort(arr.copy()))
```
